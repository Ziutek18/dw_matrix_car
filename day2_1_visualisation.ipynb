```
def groupby_and_barplot(feat_groupby, feat_agg='price_value', agg_funcs=[np.median, np.mean, np.size], feat_sort='mean', top=50, subplots=True):
  return (
    df
    .groupby(feat_groupby)[feat_agg]
    .agg(agg_funcs)
    .sort_values(by=feat_sort, ascending=False)
    .head(top)
).plot(kind='bar', figsize=(15, 5), subplots=subplots)
```

```
groupby_and_barplot('param_marka-pojazdu');
```

```
groupby_and_barplot('param_kolor', feat_sort='mean');
```
